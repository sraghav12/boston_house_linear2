Independent and dependent features


```python
X=df.drop('MEDV',axis=1)
X.head()
```

```python
Y=df['MEDV']
Y.head()
```

train test split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
X_train.shape
Y_train.shape
```

standardizing the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)   
X_test=scaler.transform(X_test)
```

```python
import pickle  
pickle.dump(scaler,open('scaling.pkl','wb'))
```

model train

```python
from sklearn.linear_model import LinearRegression
regress=LinearRegression()
```

```python
regress.fit(X_train,Y_train)
```

```python
regress.coef_
```

```python
print(regress.intercept_)
```

Parameters

```python
regress.get_params()
```

```python
# Predicting the Test set results

reg_prediction=regress.predict(X_test)
```

```python
# checking with y_test and reg_prediction

plt.scatter(reg_prediction,Y_test)
```

Well predicted

```python
# Residual(error) in reg_predictionand Y_test
residual=reg_prediction-Y_test
```

```python
sns.displot(residual,kind='kde')
```

```python
plt.scatter(reg_prediction,residual)
```

uniformly distributed

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error


print(mean_squared_error(Y_test,reg_prediction),mean_absolute_error(Y_test,reg_prediction))
```

R square

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,reg_prediction)
print(r2)
```

Adjusted R square

```python
adjusted=1-(1-r2)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
print(adjusted)
```
